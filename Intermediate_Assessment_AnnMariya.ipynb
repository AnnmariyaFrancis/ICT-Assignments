```python
train["Item_Weight"] = train["Item_Weight"].fillna(train.Item_Weight.mean())
test["Item_Weight"] = test["Item_Weight"].fillna(test.Item_Weight.mean())
```

```python
#features selecting 
```

```python
train.drop(['Item_Identifier','Outlet_Identifier'],axis=1,inplace=True)
test.drop(['Item_Identifier','Outlet_Identifier'],axis=1,inplace=True)
```

```python
train
```

```python
#preprocessing
```

```python
#label encoding
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
train['Item_Fat_Content']= le.fit_transform(train['Item_Fat_Content'])
train['Item_Type']= le.fit_transform(train['Item_Type'])
train['Outlet_Size']= le.fit_transform(train['Outlet_Size'])
train['Outlet_Location_Type']= le.fit_transform(train['Outlet_Location_Type'])
train['Outlet_Type']= le.fit_transform(train['Outlet_Type'])
```

```python
#SPLIT
```

```python
X=train.drop('Item_Outlet_Sales',axis=1)
Y=train['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=101, test_size=0.2)
```

```python
#STANDARDISE
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
X_train_std= sc.fit_transform(X_train)
X_test_std= sc.transform(X_test)
```

```python
#Logistic regression
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr.fit(X_train_std,Y_train)
```

```python
y_pred_lr=lr.predict(X_test_std)
```

```python
print(r2_score(Y_test,y_pred_lr))
print(mean_absolute_error(Y_test,y_pred_lr))
print(np.sqrt(mean_squared_error(Y_test,y_pred_lr)))
```

```python
#Random forest
```

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(n_estimators=1000)
```

```python
rf.fit(X_train_std,Y_train)
```

```python
y_pred_rf= rf.predict(X_test_std)
```

```python
print(r2_score(Y_test,y_pred_rf))
print(mean_absolute_error(Y_test,y_pred_rf))
print(np.sqrt(mean_squared_error(Y_test,y_pred_rf)))
```

```python
#hyper parameter tuning
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
# define models and parameters
model = RandomForestRegressor()
n_estimators = [10, 100, 1000]
max_depth=range(1,31)
min_samples_leaf=np.linspace(0.1, 1.0)
max_features=["auto", "sqrt", "log2"]
min_samples_split=np.linspace(0.1, 1.0, 10)
```

```python
# define grid search
grid = dict(n_estimators=n_estimators)
```

```python
#cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=101)

grid_search_forest = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, 
                           scoring='r2',error_score=0,verbose=2,cv=2)

grid_search_forest.fit(X_train_std, Y_train)
```

```python
# summarize results
print(f"Best: {grid_search_forest.best_score_:.3f} using {grid_search_forest.best_params_}")
means = grid_search_forest.cv_results_['mean_test_score']
stds = grid_search_forest.cv_results_['std_test_score']
params = grid_search_forest.cv_results_['params']

for mean, stdev, param in zip(means, stds, params):
    print(f"{mean:.3f} ({stdev:.3f}) with: {param}")
```

```python
y_pred_rf_grid=grid_search_forest.predict(X_test_std)
```

```python
# use the best model to predict on test data
y_pred_rf_grid = grid_search_forest.predict(X_test_std)

# create a new dataframe with the predicted values
df_results = pd.DataFrame({'Item_Outlet_Sales': y_pred_rf_grid})
df_results = pd.DataFrame({'Outlet_Identifier': y_pred_rf_grid})
df_results = pd.DataFrame({'Item_Outlet_Sales': y_pred_rf_grid})

# save the dataframe to a CSV file
df_results.to_csv('predicted_results.csv', index=False)
```
